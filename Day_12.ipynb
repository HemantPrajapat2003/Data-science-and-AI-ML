# Importing Modules in Python : Modules and Packages


In python , modules and packages help organize and reuse code. Here's a comprehensive guide on how to import them.

```python
import math
math.sqrt(16)
```

```python
from math import sqrt,pi
print(sqrt(16))
print(sqrt(25))
print(pi)
```

```python
import numpy as np

np.array([1,2,3,4])
```

```python
from math import * # * access all 
print(sqrt(16))
print(pi)
```

```python
from package.maths import addition
addition(2,3)
```

```python
from package import maths
from package import sub

print(maths.addition(2,3))
print(sub.subtraction(3,1))
```

## Standard Library Overview

Python's standard library is a vast collection and packages that come bundled with python , providing a wide range of functionalities out of box . Here's an overview of some of the most commonly used modules and packages in the python standard library.

```python
import array
arr = array.array("i",[1,2,3,4,5,6])
print(arr)
```

```python
import math
print(sqrt(32))
print(math.pi)
```

```python
## random

import random
print(random.randint(1,10))
print(random.choice(['apple','banana','cherry','pineapple','mango']))
```

```python
## File And Directory Acess

import os
print(os.getcwd())
```

```python
## Data Serialization 

import json
data = {'name':'Taleeb','age':25}

json_str = json.dumps(data)
print(json_str)
print(type(json_str))

parsed_data = json.loads(json_str)
print(parsed_data)
print(type(parsed_data))
```

```python
## CSV

import csv

with open('example.csv',mode='w',newline='') as file :
  writer=csv.writer(file)
  writer.writerow(['name','age'])
  writer.writerow(['taleeb',21])


with open('example.csv',mode='r') as file:
  reader=csv.reader(file)
  for row in reader:
    print(row)
```

```python
# datetime

from datetime import datetime,timedelta
now = datetime.now()

print(now)

yesterday = now - timedelta(days=1)

print(yesterday)
```

```python
## time

import time
print(time.time())
time.sleep(2)
print(time.time())
```

```python
## regular experssion

import re

pattern = r'\d+'
text='there are 123 apples 456'
match=re.search(pattern,text)
print(match.group())
```
