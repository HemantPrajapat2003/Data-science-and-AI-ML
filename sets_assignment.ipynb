```python
#assignment 2 : Add the number 11 to the set created in Assignment 1. Then remove the number 1 from the set. Print the modified set.
st = set(range(1,11))
st.add(11)
st.remove(1)
print(st)
```

```python
#assignment 3 : Create two sets: one with the first 5 positive integers and another with the first 5 even integers. Perform and print the results of union, intersection, difference, and symmetric difference operations on these sets.
set1 = set(range(1,6))
set2 = set(range(2,11,2))

union = ((set1.union(set2)))
intersection = (set1.intersection(set2))
difference = (set1.difference(set2))
sd = (set1.symmetric_difference(set2))

print(f"union : {union}")
print(f"intersection : {intersection}")
print(f"difference : {difference}")
print(f"symmetric difference : {sd}")
```

```python
# assignment 4 : Create a new set containing the squares of the first 10 positive integers using a set comprehension. Print the new set.
st = {x ** 2 for x in range (1,11)  }
print(st)
```

```python
# assignment 5 : Create a new set containing only the even numbers from the set created in Assignment 1 using a set comprehension. Print the new set.
st = { x for x in range(1,11) if  x % 2 == 0}
print(f"even numbers : {st}")
```

```python
# assignment 6 : Create a set with duplicate elements and remove the duplicates using set methods. Print the modified set.
st = { 1,2,3,4,5,6,6,7,8,8,8,9}
unique = set(st)
print(f"modified set : {unique}")
```

```python
# assignment 7 : Create two sets: one with the first 5 positive integers and another with the first 3 positive integers. Check if the second set is a subset of the first set and if the first set is a superset of the second set. Print the results.

set1 = set(range(1,6))
set2 = set(range(1,4))

print(f"set 1 : {set1}")
print(f"set 2 : {set2}")

subset = set1.issubset(set2)
superset = set1.issuperset(set2)

print(f"set 1 is subset of set 2 : {subset}")
print(f"set 1 is super set of set 2 : {superset} ")
```
