# Dictionaries

### Introduction to Dictionaries
Dictionaries are unordered collections of items. They store data in key-value pairs . Keys must be unique and immutable(e.g., strings,numbers,or tuple),while values can be of any type.

```python
## creating dictionaries 
empty_dict = {}
print(type(empty_dict))
```

```python
empty_dict = dict()
print(empty_dict)
```

```python
students = {"name":"taleeb", "roll:NO":31, "age":21}
print(students)
print(type(students))
```

```python
## single key is always used 
students = {"name":"taleeb", "roll:NO":31, "name":21} # only unique key lga 
print(students)
```

```python
## Acessing Dictionary elements 
students = {"name":"taleeb", "roll:NO":31, "age":21}
print(students['age'])
print(students['name'])
```

```python
# Accesing usning get() method
students = {"name": "taleeb","age":21,"grade":"A"}
print(students.get("name"))
print(students.get('grade'))
print(students.get("last name","Not available"))
```

```python
## modified dictonary elements 
#Dictionary are mutable , so you can add , update or delete elements .
students = {"name": "taleeb","age":21,"grade":"A"}

students['grade']="A++"# updated the value for the key
print(students)
students["address"]="rajasthan"# create new key value 
print(students)
del students["grade"]# delete key and value pair
print(students)
```

```python
## Dictionary methods

keys=students.keys() # gets all the keys
print(keys)
values=students.values() # gets all values
print(values)
items = students.items() #get all keys pairs
print(items)
```

```python
### shallow copy

student_copy = students
print(students)
print(student_copy)

students["name"] = "taleeb2"
print(students)
print(student_copy)
```

```python
student_copy1 = students.copy() # shallow copy
print(student_copy1)
print(students)

students["name"] = "taleeb3"
print(students)
print(student_copy1)
```

```python
# iterating over Dictionaries
## you can use loops to iterate over dictionatries , keys , values  or items


## iterating over keys
for keys in students.keys():
  print(keys)
```

```python
# iterate over values
for values in students.values():
  print(values)
```

```python
## iterate over key value pairs
for key, value in students.items():
  print(f"{key}:{value}")
```

```python
## Nested dictionaries

students ={
  "student1":{"name":"taleeb","age":21,},
  "student2":{"name":"danish","age":22},
  "student3":{"name":"kaif","age":20},
  "student4":{"name":"aziz","age":19}
}

print(students)
```

```python
## Acess nested dictionaries elements
print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
print(students.items())
```

```python
# iterating over nested dictionaries
students ={
  "student1":{"name":"taleeb","age":21,},
  "student2":{"name":"danish","age":22},
  "student3":{"name":"kaif","age":20},
  "student4":{"name":"aziz","age":19}
}

for student_id , student_info in students.items():
  print(f"{student_id}:{student_info}")
  for key , value in student_info.items():
    print(f"{key}:{value}")
```

```python
## Dictionary Compherension
 
squares = {x:x**2 for x in range(5)}
print(squares)
```

```python
## condition dictionary compherension

even = {x:x**2 for x in range(10) if x %2 == 0}
print(even)
```

```python
# practical examples

## use a dict to count the frequency of element in list

numbers = [1,2,2,3,3,3,4,4,4,4]
frequency ={}

for number in numbers:
  if number in frequency:
    frequency[number]+=1
  else:
    frequency[number]=1
print(frequency)
```

```python
# Merge 2 dict into one


dict1 = {"a":1,"b":2}
dict2 = {"c":3,"d":4}
merged_dict = {**dict1,**dict2}
print(merged_dict)
```
