# Real-World Examples Using List in Python

List are one of the most commonly used data structures in python. thanks to their versatility and ease to use. 

Here are serveral practical examples that illustrate their use in real-world scenarios.

## Example 1. Manage a TO do List

### > Create a To do list keep track of tasks 


```python
to_do_list = ["buy grocesries","clean the house","pay bills"]

# adding to task
to_do_list.append("Schedule meeting")
to_do_list.append("go for run")

## removing a complete task
to_do_list.remove("clean the house")

## checking if a task is in the list
if "pay bills" in to_do_list:
  print("Don't forget to pay the utility bills ")

print("To do List remaining")
for task in to_do_list:
  print(f"-{task}")
```

## Example 2:Organizing Students Grades

### > create a list to store and calculate average grades for students

```python
## Oraganizing student grades
grades = [85,92,78,90,88,95,65]

## adding a new grade
grades.append(55)

# calculating the average grade
average_grade = sum(grades) / len(grades)
print(f"Average Grades : {average_grade:.2f}")

#finding the highest and lowest grades
highest_grade = max(grades)
lowest_grade = min(grades)
print(f"highest grade : {highest_grade}")
print(f"lowest grades : {lowest_grade}")
```

## Example 3: Managing An Inventory

### > Use a list to collect and analyze usee feedback

```python
# manage an inventory 
inventory = ["apples","bananas","oranges","grapes"]


#adding a new items
inventory.append("strawberries")

#removing an item that is out of stock
inventory.remove("bananas")

#checking if an item is in stock
item = "oranges"
if item in inventory:
  print(f"{item} in stock.")
else:
  print(f"{item} are out of stock.")

## printing the inventory
print("Inventory List:")
for item in inventory:
  print(f"-{item}")
```

## Example 4: Collecting User feedback

### > Use a list to collect and analyze user feedback.

```python
## collecting user feedback
feedback = ["Great service!","Very satisfied", "Could be better","Excellent experince"]

#adding new feedback
feedback.append("not happy with the service")

# counting specific feedback
positve_feedbac_count = sum(1 for comment in feedback if "Great" in comment.lower() or "excellent" in comment.lower())
print(f"Positive Feedback Count : {positve_feedbac_count}")

#printing all feedback
print("User feedback:")
for comment in feedback:
  print(f"-{comment}")
```
