# File Operation - Read And Write Files

File handling is a crucial part of any programming language. Python provides built-in fucntions and methods to read from and write to files , both text and binary. This lesson will cover the basics of file handling , including reading and writing text files and binary files.

```python
## read a whole file 

with open('example.txt', 'r') as file:
  content = file.read()
  print(content)
```

```python
## Read a file line by line

with open('example.txt','r') as file:
  for line in file:
    print(line.strip()) # strip() remove the newline character
```

```python
## Writing a file (overwriting)

with open('example.txt','w') as file:
  file.write('Hello world! \n')
  file.write('this is a new line.')
```

```python
## Write a file (without overwriting)

with open('example.txt','a') as file:
  file.write("Append operation taking place! \n")
  file.write("Append operation taking place!\n")

```

```python
## writing a list of lines to a file

lines =['First line \n','Second line \n','Third line\n']

with open('example.txt','a') as file:
  file.writelines(lines)
```

```python
### Binary Files
 
## Writing to a binary File
data = b'\x00\x01\x02\x03\x04'
with open('example.bin','wb') as file:
  file.write(data)
```

```python
## Reading a binary file 
with open('example.bin','rb') as file:
  content = file.read()
  print(content)
```

```python
## read the content from source text file and write to a destination text file 

# Copying a text file
with open('example.txt','r') as source_file:
  content = source_file.read()

with open('destination.txt','w') as destination_file:
  destination_file.write(content)
```

The w+ mode in python is used to open a file for both reading and writing . If the file does not exist , it will br created . If the file exists , its content is truncated (i.e , the file is overwritten) 

```python
## writing and then reading a file

with open('example.txt','w+') as file :
  file.write("Hello world \n")
  file.write("this is new line\n")

  ## Move the file cursor to the beginning
  file.seek(0)

  ## read the content of the file
  content=file.read()


  print(content)
```
